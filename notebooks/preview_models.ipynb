# 3D Models Preview

`models/` 디렉토리 하위에 있는 STEP 파일들을 모두 불러와서 OCP 뷰어에서 렌더링합니다.

```python
import os
from pathlib import Path

import jupyter_cadquery
from jupyter_cadquery import show

from build123d import import_step
```

```python
# 모델 탐색 및 로드
models_dir = Path("models")
step_files = list(models_dir.rglob("*.step"))

loaded_models = {}
for f in step_files:
    print(f"Loading {f.name}...")
    try:
        model = import_step(str(f))
        loaded_models[f.stem] = model
    except Exception as e:
        print(f"Failed to load {f.name}: {e}")

print(f"\n총 {len(loaded_models)}개의 모델을 로드했습니다.")
```

```python
# 모든 모델을 순차적으로 보여주기
if loaded_models:
    for name, model in loaded_models.items():
        print(f"\n--- {name} ---")
        show(model)
else:
    print("표시할 모델이 없습니다.")
```

### 개별 모델 뷰어
특정 모델만 보고 싶다면 아래 셀에서 이름을 지정해 실행하세요.

```python
if loaded_models:
    # 첫 번째 모델 이름 가져오기
    first_model_name = list(loaded_models.keys())[0]
    print(f"Showing {first_model_name}...")
    
    # jupyter_cadquery show() 사용
    show(loaded_models[first_model_name])
```
